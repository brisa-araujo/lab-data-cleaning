#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
#defino mis variables "resumen" con solamente las columnas que interesan.
#.shape para saber cuántas líneas y columnas tengo
users_summary = users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
posts_summary = posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
print(f'Shape de users_summary: {users_summary.shape}')
print(f'Shape de posts_summary: {posts_summary.shape}')
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
#vimos en el shape que el df de posts es más extenso que users y que ambos tienen el mismo número de columnas.
#sin embargo, no tienen las MISMAS columnas; representan información distinta.
#mi merge tendría entonces que repetir información de user para cada post, de forma que el único id único de la tabla sea el de postId
#para esto, realizo un merge con parámetro "right" y variable base 'userId', análogo a hacer un RIGHT JOIN en sql.
merge_df = users_summary.merge(posts_summary, how='right', on='userId')
merge_df.head(10)
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
#Primer paso: calcular IQR - interquartile range, rango entre 1er y 3er cuartiles
stats = merge_df.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
#Segundo paso: crear df vacío con las mismas columnas de merge_df
outliers = pd.DataFrame(columns=merge_df.columns)

#Tercer paso: identificar outliers en cada columna del df original, haciendo loop en el df para
#a) Calcular los límites bajo y alto
#b) Recuperar los registros que están fuera de los límites
#c) Guardar esos resultados en el nuevo df de outliers

for col in stats.index: #cada index de stats tiene el nombre de una columna del df
    iqr = stats.at[col,'IQR'] #recupero el nombre de la columna y el valor del IQR
    cutoff = iqr * 1.5 #establezco el límite multiplicando el IQR * 1.5
    lower = stats.at[col,'25%'] - cutoff #límite bajo es 1er cuartil - límite
    upper = stats.at[col,'75%'] + cutoff #límite alto es 3er cuartil + límite
    results = merge_df[(merge_df[col] < lower) | 
                   (merge_df[col] > upper)].copy() #si valor del df es menor que el límite bajo o mayor que límite alto, lo guardo
    results['Outlier'] = col #agrego columna de outlier a ese resultado guardado con el nombre de la variable en el que es outlier
    outliers = outliers.append(results) #agrego esta columna guardada a mi df de outliers

print(outliers['Outlier'].value_counts())

#Guardo mis outliers en csv para enviar
outliers.to_csv('./outliers.csv', index=False)
```
